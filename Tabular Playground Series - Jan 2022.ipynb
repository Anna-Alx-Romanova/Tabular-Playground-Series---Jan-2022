```python
train_encoded = pd.get_dummies(train, columns = ['country' , 'store' , 'product']) 
train_encoded.head()
```

```python
test_encoded = pd.get_dummies(test, columns = ['country' , 'store' , 'product']) 
test_encoded.head()
```

```python
train_encoded.index=train_encoded['date']
test_encoded.index=test_encoded['date']
train_encoded.head()
```

```python
train_encoded=train_encoded.drop('date',axis=1)
test_encoded=test_encoded.drop('date',axis=1)
```

```python
train_encoded=train_encoded.drop('row_id',axis=1)
```

```python
X=train_encoded.drop('num_sold',axis=1)
y=train_encoded['num_sold']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
y_valid
```

```python
from catboost import CatBoostRegressor
cat = CatBoostRegressor()
cat.fit(X_train,y_train)
```

```python
X_check=cat.predict(X_valid)
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(X_check,y_valid))
```
